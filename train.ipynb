```python
import datasets
from datasets import load_dataset

cv_16 = load_dataset("mozilla-foundation/common_voice_16_1", "yue", split="train")
```

```python
import os
import torch
import torchaudio

output_dir = 'train_wavs'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def convert_mp3_to_wav(sample):
    path = os.path.splitext(os.path.basename(sample['path']))[0]
    audio = sample['audio']
    wav_path = f"{output_dir}/{path}.wav"

    target_sampling_rate = 16000

    # Ensure audio data is a PyTorch tensor in the correct shape (channels, frames)
    if not isinstance(audio['array'], torch.Tensor):
        audio_data = torch.tensor(audio['array'], dtype=torch.float32).unsqueeze(0)  # Adding channel dimension for mono audio
    else:
        audio_data = audio['array'].unsqueeze(0) if audio_data.dim() == 1 else audio_data

    # Resample to 16K Hz if necessary
    if audio['sampling_rate'] != target_sampling_rate:
        resampler = torchaudio.transforms.Resample(orig_freq=audio['sampling_rate'], new_freq=target_sampling_rate)
        audio_data = resampler(audio_data)
    
    # Ensure audio data is in the correct shape and type for saving
    if audio_data.dim() == 1:  # If the audio is mono, add an extra dimension
        audio_data = audio_data.unsqueeze(0)

    # Convert mp3 to wav and save
    torchaudio.save(wav_path, audio_data, target_sampling_rate)

# Convert and save the mp3 files as wav in parallel, ensuring 16K Hz sampling rate
cv_16.map(convert_mp3_to_wav)
```

```python
import re

def process_sentence(sample):
    # remove punctuation
    sample['sentence'] = re.sub(r'[^\u4e00-\u9FFFa-zA-Z0-9 ]', '', sample['sentence'])
    # add space between Chinese characters
    sample['sentence'] =re.sub(r'([\u4e00-\u9fff])', r'\1 ', sample['sentence']).strip()
    # add space after an English word followed by a Chinese character
    sample['sentence'] = re.sub(r'([a-zA-Z0-9_]+)([\u4e00-\u9fff])', r'\1 \2', sample['sentence'])
    # Condense consecutive whitespaces into one
    sample['sentence'] = re.sub(r'\s+', ' ', sample['sentence']).strip()
    return sample

processed_cv_16 = cv_16.map(process_sentence)
```

```python
from praatio import textgrid

for sample in processed_cv_16:
    try:
        audio_info = torchaudio.info(sample['audio']['path'])
        duration_seconds = audio_info.num_frames / audio_info.sample_rate
        tg_path = os.path.join(output_dir, os.path.splitext(os.path.basename(sample['path']))[0] + '.TextGrid')
        entry = (0, duration_seconds, sample['sentence'])
        #print(entry)
        wordTier = textgrid.IntervalTier(sample['client_id'], [entry], 0, duration_seconds)
        tg = textgrid.Textgrid()
        tg.addTier(wordTier)
        tg.save(tg_path, format="short_textgrid", includeBlankSpaces=True)
    except Exception as e:
        print("Failed to write file",e)
```

# Build lexicon

```python
all_characters = set()
for sample in processed_cv_16:
    all_characters.update(set(sample['sentence'].replace(' ', '')))

all_characters = list(all_characters)
```

```python
all_characters.sort()
print(len(all_characters))
```

```python
import pandas as pd

char_df = pd.read_csv('char.csv')
filtered_chars = char_df[(char_df['char'].isin(all_characters)) & (char_df['pron_rank'].isin(['預設', '常用']))]
print(len(filtered_chars))
```

```python
import pycantonese

def jyutping_to_phonemes(jyutping):
    syllables = pycantonese.parse_jyutping(jyutping)
    phonemes = []
    for syllable in syllables:
        if syllable.onset != '':
            phonemes.append(syllable.onset)
        if syllable.nucleus != '':
            phonemes.append(syllable.nucleus + syllable.tone)
        if syllable.coda != '':
            phonemes.append(syllable.coda)
    return phonemes

assert(jyutping_to_phonemes('gwai2') == ['gw', 'a2', 'i'])
assert(jyutping_to_phonemes('m4') == ['m4'])

with open('lexicon.txt', 'w', encoding='utf-8') as f:
    f.write('{LG}\tspn\n')
    f.write('{SL}\tsil\n')
    for index, row in filtered_chars.iterrows():
        char = row['char']
        jyutping = row['jyutping']
        f.write(f"{char}\t{' '.join(jyutping_to_phonemes(jyutping))}\n")
```
