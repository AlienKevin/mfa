```python
import os
import torch
import torchaudio

output_dir = 'train_wavs'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def convert_mp3_to_wav(sample):
    path = os.path.splitext(os.path.basename(sample['path']))[0]
    audio = sample['audio']
    wav_path = f"{output_dir}/{path}.wav"

    target_sampling_rate = 16000

    # Ensure audio data is a PyTorch tensor in the correct shape (channels, frames)
    if not isinstance(audio['array'], torch.Tensor):
        audio_data = torch.tensor(audio['array'], dtype=torch.float32).unsqueeze(0)  # Adding channel dimension for mono audio
    else:
        audio_data = audio['array'].unsqueeze(0) if audio_data.dim() == 1 else audio_data

    # Resample to 16K Hz if necessary
    if audio['sampling_rate'] != target_sampling_rate:
        resampler = torchaudio.transforms.Resample(orig_freq=audio['sampling_rate'], new_freq=target_sampling_rate)
        audio_data = resampler(audio_data)
    
    # Ensure audio data is in the correct shape and type for saving
    if audio_data.dim() == 1:  # If the audio is mono, add an extra dimension
        audio_data = audio_data.unsqueeze(0)

    # Convert mp3 to wav and save
    torchaudio.save(wav_path, audio_data, target_sampling_rate)

# Convert and save the mp3 files as wav in parallel, ensuring 16K Hz sampling rate
cv_16.map(convert_mp3_to_wav)
```

```python
import re

def process_sentence(sample):
    # remove punctuation
    sample['sentence'] = re.sub(r'[^\u4e00-\u9fff\u3400-\u4dbf\U00020000-\U0002a6df\U0002a700-\U0002ebef\U00030000-\U000323af\ufa0e\ufa0f\ufa11\ufa13\ufa14\ufa1f\ufa21\ufa23\ufa24\ufa27\ufa28\ufa29\u3006\u3007a-zA-Z0-9 ]', '', sample['sentence'])
    # add space between Chinese characters
    sample['sentence'] =re.sub(r'([\u4e00-\u9fff\u3400-\u4dbf\U00020000-\U0002a6df\U0002a700-\U0002ebef\U00030000-\U000323af\ufa0e\ufa0f\ufa11\ufa13\ufa14\ufa1f\ufa21\ufa23\ufa24\ufa27\ufa28\ufa29\u3006\u3007])', r'\1 ', sample['sentence']).strip()
    # add space after an English word followed by a Chinese character
    sample['sentence'] = re.sub(r'([a-zA-Z0-9_]+)([\u4e00-\u9fff])', r'\1 \2', sample['sentence'])
    # Condense consecutive whitespaces into one
    sample['sentence'] = re.sub(r'\s+', ' ', sample['sentence']).strip()
    return sample

processed_cv_16 = cv_16.map(process_sentence)
```

```python
from praatio import textgrid

for sample in processed_cv_16:
    try:
        audio_info = torchaudio.info(sample['audio']['path'])
        duration_seconds = audio_info.num_frames / audio_info.sample_rate
        tg_path = os.path.join(output_dir, os.path.splitext(os.path.basename(sample['path']))[0] + '.TextGrid')
        entry = (0, duration_seconds, sample['sentence'])
        #print(entry)
        wordTier = textgrid.IntervalTier(sample['client_id'], [entry], 0, duration_seconds)
        tg = textgrid.Textgrid()
        tg.addTier(wordTier)
        tg.save(tg_path, format="short_textgrid", includeBlankSpaces=True)
    except Exception as e:
        print("Failed to write file",e)
```

# Build lexicon

```python
all_characters = set()
for sample in processed_cv_16:
    all_characters.update(set(sample['sentence'].replace(' ', '')))

all_characters = list(all_characters)
```

```python
import pandas as pd

all_words = pd.read_csv('word.csv')
all_words = all_words['char'].tolist()

print(len(all_words))
```

```python
import ahocorasick
from tqdm import tqdm

# Create the Aho-Corasick automaton
A = ahocorasick.Automaton()

# Add words to the automaton
for idx, word in enumerate(all_words):
    A.add_word(word, (idx, word))

# Prepare the automaton
A.make_automaton()

# Found words set
found_words = set()

# Search the corpus
for sentence in tqdm(processed_cv_16['sentence'], desc="Processing sentences"):
    # If you need to remove spaces or preprocess each sentence, do it here
    processed_sentence = sentence.replace(' ', '')
    for end_index, (idx, original_word) in A.iter(processed_sentence):
        found_words.add(original_word)

print(len(found_words))
```

```python
print(f"First 10 words found: {list(found_words)[:10]}")
```

```python
import pandas as pd

words_to_jyutpings = {}

for index, entry in pd.read_csv('word.csv').iterrows():
    if entry['char'] in found_words:
        words_to_jyutpings[entry['char']] = entry['jyutping'] 

print(len(words_to_jyutpings))
```

```python
import re

chinese_char_pattern = re.compile(r'[\u4e00-\u9fff\u3400-\u4dbf\U00020000-\U0002a6df\U0002a700-\U0002ebef\U00030000-\U000323af\ufa0e\ufa0f\ufa11\ufa13\ufa14\ufa1f\ufa21\ufa23\ufa24\ufa27\ufa28\ufa29\u3006\u3007]')
```

```python
from collections import defaultdict

char_jyutpings_from_words = defaultdict(set)

for word, jyutping in words_to_jyutpings.items():
    chars = [char for char in word if chinese_char_pattern.match(char)]
    jyutpings = jyutping.split(" ")
    if len(chars) != len(jyutpings):
        print(f"Word: {word}, Chars: {chars}, Jyutpings: {jyutpings}")
    else:
        for char, jyutping in zip(chars, jyutpings):
            if char == '心' and jyutping == 'jan4':
                print(f"Word: {word}, Char: {char}, Jyutping: {jyutping}")
            char_jyutpings_from_words[char].add(jyutping)

print(len(char_jyutpings_from_words))
print("First 10 entries in the dictionary:")
print(dict(list(char_jyutpings_from_words.items())[:10]))
```

```python
import pandas as pd

char_df = pd.read_csv('char.csv')
filtered_chars = char_df[(char_df['char'].isin(all_characters)) & (char_df['pron_rank'].isin(['預設', '常用']))]
print(len(filtered_chars))

# Convert the dataframe to a dictionary
filtered_char_jyutpings = defaultdict(set)
for index, row in filtered_chars.iterrows():
    char = row['char']
    jyutping = row['jyutping']
    filtered_char_jyutpings[char].add(jyutping)

print(len(filtered_char_jyutpings))
```

```python
new_char_jyutpings = defaultdict(set)

# Print jyutpings from words that are not present in char.csv
for char, jyutpings in char_jyutpings_from_words.items():
    for jyutping in jyutpings:
        if jyutping not in filtered_char_jyutpings[char]:
            new_char_jyutpings[char].add(jyutping)

print(len(new_char_jyutpings))
print("First 10 entries in the new_char_jyutpings:")
print(dict(list(new_char_jyutpings.items())[:10]))

# print any new characters that are not in the char.csv
new_characters = set(new_char_jyutpings.keys()) - set(filtered_char_jyutpings.keys())
print(f"New characters: {new_characters}")
```

```python
# Merge the new jyutpings with the filtered_char_jyutpings
for char, jyutpings in new_char_jyutpings.items():
    filtered_char_jyutpings[char] |= jyutpings

print(len(filtered_char_jyutpings))
```

```python
import pycantonese

def jyutping_to_phonemes(jyutping):
    syllables = pycantonese.parse_jyutping(jyutping)
    phonemes = []
    for syllable in syllables:
        if syllable.onset != '':
            phonemes.append(jyutping_onset_to_ipa(syllable.onset))
        if syllable.nucleus != '':
            coda = ('' if (syllable.coda in ['i', 'u']) else syllable.coda)
            nucleus = jyutping_nucleus_to_ipa(syllable.nucleus + (syllable.coda if syllable.coda in ['i', 'u'] else ''), coda)
            tone = jyutping_tone_to_ipa(syllable.tone)
            phonemes.append(nucleus + tone)
        if syllable.coda != '':
            phonemes.extend([] if syllable.coda in ['i', 'u'] else [jyutping_coda_to_ipa(syllable.coda)])
    return phonemes

def jyutping_onset_to_ipa(onset):
    if onset == 'b': return 'p'
    if onset == 'c': return 'tsʰ'
    if onset == 'd': return 't'
    if onset == 'g': return 'k'
    if onset == 'gw': return 'kʷ'
    if onset == 'k': return 'kʰ'
    if onset == 'kw': return 'kʷʰ'
    if onset == 'ng': return 'ŋ'
    if onset == 'p': return 'pʰ'
    if onset == 't': return 'tʰ'
    if onset == 'z': return 'ts'
    return onset

def jyutping_nucleus_to_ipa(nucleus, coda):
    if nucleus == 'a': return 'ɐ'
    if nucleus == 'aa': return 'aː'
    if nucleus == 'aai': return 'aːj'
    if nucleus == 'aau': return 'aːw'
    if nucleus == 'ai': return 'ɐi'
    if nucleus == 'au': return 'ɐw'
    if nucleus == 'e': return 'ɛː'
    if nucleus == 'ei': return 'ej'
    if nucleus == 'eo': return 'ɵ'
    if nucleus == 'eoi': return 'ɵɥ'
    if nucleus == 'eu': return 'ɛːw'
    if nucleus == 'i': return 'ɪ' if coda in ['k', 'ng'] else 'iː'
    if nucleus == 'iu': return 'iːw'
    if nucleus == 'o': return 'ɔː'
    if nucleus == 'oe': return 'œː'
    if nucleus == 'oi': return 'ɔːj'
    if nucleus == 'ou': return 'ow'
    if nucleus == 'u': return 'ʊ' if coda in ['k', 'ng'] else 'uː'
    if nucleus == 'ui': return 'uːj'
    if nucleus == 'yu': return 'yː'
    if nucleus == 'm': return 'm̩'
    if nucleus == 'ng': return 'ŋ̍'
    return nucleus

def jyutping_coda_to_ipa(coda):
    if coda == 'p': return 'p̚'
    if coda == 't': return 't̚'
    if coda == 'k': return 'k̚'
    if coda == 'ng': return 'ŋ'
    return coda

def jyutping_tone_to_ipa(tone):
    if tone == '1': return '˥'
    if tone == '2': return '˨˥'
    if tone == '3': return '˧'
    if tone == '4': return '˨˩'
    if tone == '5': return '˨˧'
    if tone == '6': return '˨'
    raise ValueError(f"Invalid tone input {tone}")


assert(jyutping_to_phonemes('gwai5') == ['kʷ', 'ɐi˨˧'])
assert(jyutping_to_phonemes('m4') == ['m̩˨˩'])
assert(jyutping_to_phonemes('ng5') == ['ŋ̍˨˧'])
assert(jyutping_to_phonemes('noeng4') == ['n', 'œː˨˩', 'ŋ'])
assert(jyutping_to_phonemes('zoi2') == ['ts', 'ɔːj˨˥'])
assert(jyutping_to_phonemes('cyun1') == ['tsʰ', 'yː˥', 'n'])
assert(jyutping_to_phonemes('ceon4') == ['tsʰ', 'ɵ˨˩', 'n'])
assert(jyutping_to_phonemes('heoi5') == ['h', 'ɵɥ˨˧'])
assert(jyutping_to_phonemes('kwok3') == ['kʷʰ', 'ɔː˧', 'k̚'])
assert(jyutping_to_phonemes('sing1') == ['s', 'ɪ˥', 'ŋ'])
assert(jyutping_to_phonemes('zik6') == ['ts', 'ɪ˨', 'k̚'])
assert(jyutping_to_phonemes('ziu1') == ['ts', 'iːw˥'])
assert(jyutping_to_phonemes('nguk1') == ['ŋ', 'ʊ˥', 'k̚'])
assert(jyutping_to_phonemes('fut3') == ['f', 'uː˧', 't̚'])
assert(jyutping_to_phonemes('zau1') == ['ts', 'ɐw˥'])
assert(jyutping_to_phonemes('lei5') == ['l', 'ej˨˧'])
assert(jyutping_to_phonemes('ngong6') == ['ŋ', 'ɔː˨', 'ŋ'])
assert(jyutping_to_phonemes('ngung6') == ['ŋ', 'ʊ˨', 'ŋ'])
assert(jyutping_to_phonemes('aa1') == ['aː˥'])
assert(jyutping_to_phonemes('aai3') == ['aːj˧'])
assert(jyutping_to_phonemes('aau1') == ['aːw˥'])

with open('lexicon.txt', 'w', encoding='utf-8') as f:
    f.write('{LG}\tspn\n')
    f.write('{SL}\tsil\n')
    for char, jyutpings in filtered_char_jyutpings.items():
        for jyutping in sorted(jyutpings):
            f.write(f"{char}\t{' '.join(jyutping_to_phonemes(jyutping))}\n")
```
